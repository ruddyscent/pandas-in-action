```python
cars = pd.read_csv("used_cars.csv")
cars.head()
```

```python
min_wage = pd.read_csv("minimum_wage.csv")
min_wage.head()
```

# 1

```python
cars.pivot_table(index="Fuel", values="Price", aggfunc="sum")
```

# 2

```python
cars.pivot_table(index="Manufacturer", columns="Transmission", values="Price", aggfunc="sum", margins=True)
```

# 3

```python
cars.pivot_table(index=["Year", "Fuel"], columns="Transmission", values="Price", aggfunc="mean")
```

# 4

```python
cars.pivot_table(index=["Year", "Fuel"], columns="Transmission", values="Price", aggfunc="mean").stack()
```
