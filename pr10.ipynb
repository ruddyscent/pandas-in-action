```python
week1 = pd.read_csv("week_1_sales.csv")
week2 = pd.read_csv("week_2_sales.csv")
customers = pd.read_csv("customers.csv", index_col="ID")
foods = pd.read_csv("foods.csv", index_col="Food ID")
```

# 1

```python
pd.concat([week1, week2], keys=["Week 1", "Week 2"])
```

# 2

```python
week1.merge(week2, how="inner", on="Customer ID")
```

# 3

```python
week1.merge(week2, how="inner", on=["Customer ID", "Food ID"])
```

# 4

```python
outer_join = week1.merge(week2, how="outer", on="Customer ID", indicator=True)
outer_join
```

```python
outer_join[outer_join["_merge"] == "left_only"]
```

```python
outer_join[outer_join["_merge"] == "right_only"]
```

# 5

```python
week1.merge(customers, how="left", left_on="Customer ID", right_index=True)
```
